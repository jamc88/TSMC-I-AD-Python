El objeto **TextFileReader** devuelto por pandas.read_csv permite iterar a lo largo de las partes del archivo segn los recuentos de valor de la columna deseada.

```python
df = pd.read_csv('victimasFGJ2023.csv', chunksize=1000)

tot = pd.Series([], dtype='int64')
    
for piece in df:
    tot = tot.add(piece['delito'].value_counts(), fill_value=0)
    tot = tot.sort_values(ascending=False)
```

```python
tot
```

```python
delitos = pd.DataFrame(tot,columns=['Frecuencia'])
```

```python
delitos.head(20)
```

```python
delitos = delitos.head(20).sort_values(by=['Frecuencia'])
```

```python
plt.figure(figsize=(7,9), dpi=150)
plt.barh(delitos.index, delitos['Frecuencia'],facecolor='#6495ED', edgecolor='white')
plt.title('Delitos CDMX', fontsize = 15)
plt.show()





```

Escribir datos en formato de texto

```python
delitos.to_csv('delitos.csv') # sep='|'
```

No habiendo otras opciones especificadas, se escriben tanto las etiquetas de fila como de columna, y ambas se pueden deshabilitar
