```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df20 = pd.read_csv("https://raw.githubusercontent.com/jamc88/TSMCI-Analisis-de-datos-con-Python/main/Datos/homicidios_2020.csv")
```

```python
df20
```

```python
data = pd.DataFrame({'Homicidios' : [df20['Mujeres'].sum(),
                                     df20['Hombres'].sum(),
                                     df20['No_esp'].sum()]},
                    index=('Mujeres','Hombres','No especificado'))

```

```python
data
```

```python
plt.figure(figsize=(7,4), dpi=250) 
plt.bar(data.index, data['Homicidios'],color='#008B8B')
plt.title('Total de homicidios por género durante 2020')
plt.show()
```

```python
plt.barh(data.index, data['Homicidios'])
plt.title('Total de homicidios por género durante 2020')
plt.show()
```

# Homicidios en México (2020-2022)

```python
df21 = pd.read_csv("https://raw.githubusercontent.com/jamc88/TSMCI-Analisis-de-datos-con-Python/main/Datos/homicidios_2021.csv")
df22 = pd.read_csv("https://raw.githubusercontent.com/jamc88/TSMCI-Analisis-de-datos-con-Python/main/Datos/homicidios_2022.csv")
```

```python
H = [sum(df20['Hombres']),sum(df21['Hombres']),sum(df22['Hombres'])]
M = [sum(df20['Mujeres']),sum(df21['Mujeres']),sum(df22['Mujeres'])]
NE = [sum(df20['No_esp']),sum(df21['No_esp']),sum(df22['No_esp'])]
```

```python
data_Y = pd.DataFrame({'Hombres' : H,
                       'Mujeres' : M,
                       'No_especificado': NE},
                    index=('2020','2021','2022'))
```

```python
data_Y
```

```python
plt.bar(data_Y.index, data_Y['Mujeres'],color='orange')
plt.title('Total de homicidios de Mujeres por año')
plt.show()
```

Hombres, NE ...

```python
n = len(data_Y.index)
x = np.arange(n)
width = 0.25
plt.bar(x - width, data_Y.Hombres, width=width, label='Hombres')
plt.bar(x, data_Y.Mujeres, width=width, label='Mujeres')
plt.bar(x + width, data_Y.No_especificado, width=width, label='No especificado')
plt.xticks(x, data_Y.index)
plt.legend(loc=1)

plt.show()
```

# Tasa de Homicidios por cada 100,000 habitantes (2020)

$$
Tasa \ de\ homicidios = \left( \frac{Total\ de \ homicidios\ en\ el\ año}{Población \ Total} \right)\times 100,000
$$

```python
df_poblacion = pd.read_csv("https://raw.githubusercontent.com/jamc88/TSMCI-Analisis-de-datos-con-Python/main/Datos/poblacion.csv")
```

```python
df_poblacion
```

```python
df20
```

```python
df20['Poblacion'] = df_poblacion['H_2020'] + df_poblacion['M_2020']
```

```python
df20
```

```python
df20
```

```python
df20['Tasa'] = (df20['Total']/df20['Poblacion'])*100000
```

```python
df20
```

```python
df20.sort_values(by=['Tasa'], ascending=False)
```

```python
plt.figure(figsize=(7,9), dpi=150)
hom = df20.sort_values(by=['Tasa'])



plt.barh(hom['Estado'], hom['Tasa'],facecolor='#9999ff', edgecolor='white')
plt.title('Tasa de homicidios por estado durante 2020', fontsize = 15)
plt.show()
```
