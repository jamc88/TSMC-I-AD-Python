# Indexación jerárquica

La indexación jerárquica es una característica importante de pandas que permite tener varios niveles de índice en un eje o, dicho de otro modo, proporciona una forma de trabajar con datos de muchas dimensiones en una forma de menor dimensión.

```python
df = pd.read_csv('https://raw.githubusercontent.com/jamc88/TSMCI-Analisis-de-datos-con-Python/main/Datos/ingresosstc_2023_07.csv', index_col=['fecha']) 
```

```python
df
```

```python
M21 = df.loc['2021-01-01':'2021-12-31']
```

```python
M21
```

```python
dat = M21.groupby(['linea','tipo_ingreso']).sum()
dat
```

```python
dat.index
```

```python
dat.loc['Línea 3']
```

```python
dat.loc['Línea 3':'Línea 5']
```

La indexación jerárquica desempeña un papel de gran importancia en la remodelación de datos y en operaciones basadas en grupos, como, por ejemplo, en la formación de una tabla dinámica.

Se pueden reordenar estos datos en un dataframe utilizando su método **unstack**:

```python
dat.unstack()
```

La operación inversa a unstack es **stack**:

```python
dat.unstack().stack()
```

Se puede saber cuantos niveles tiene un índice accediendo a su atributo **nlevels**:

```python
dat.index.nlevels
```

**sort_index** clasifica los datos lexicográficamente, utilizando todos los niveles de índice

```python
dat.sort_index(level=1)
```

Se puede tener la opción de emplear solamente un nivel o un subconjunto de niveles para realizar la clasificación pasando el argumento level. 

```python
dat.swaplevel(0, 1).sort_index(level=0)
```

Muchas estadísticas descriptivas y de resumen en objetos DataFrame y Series tienen una opción level.

```python
dat.groupby(level='tipo_ingreso').sum()
```

```python
dat.groupby(level='tipo_ingreso').sum()
```

# Uniones de dataframes

Las operaciones de fusión o unión combinan conjuntos de datos enlazando filas mediante una o varias claves. 

```python
df1 = pd.read_csv('https://raw.githubusercontent.com/jamc88/TSMCI-Analisis-de-datos-con-Python/main/Datos/nacimientos_2021.csv',nrows=5, usecols=['Estado','Regiones','Hombres','Mujeres']) 
df2 = pd.read_csv('https://raw.githubusercontent.com/jamc88/TSMCI-Analisis-de-datos-con-Python/main/Datos/nacimientos_2022.csv',nrows=8, usecols=['Estado','Hombres','Mujeres'])
```

```python
df2
```

**pandas.merge:** Conecta filas de dataframes según una o varias claves.


```python
pd.merge(df1,df2, on = 'Estado')
```

La unión externa toma la unión de las claves.

```python
pd.merge(df1, df2, how='outer')
```

```python
pd.merge(df1, df2, on='Estado', how='outer')
```

```python
c19 = pd.read_excel('contam2019.xlsx', usecols=['Fecha','Centro Ozono']) 
c20 = pd.read_excel('contam2020.xlsx', usecols=['Fecha','Centro Ozono'])
c21 = pd.read_excel('contam2021.xlsx', usecols=['Fecha','Centro Ozono'])
```

```python
df = pd.merge(c19, c20, how='outer')
```

```python
df
```

```python
df['Fecha'] = pd.to_datetime(df['Fecha'])
df = df.set_index("Fecha")
```

```python
plt.figure(figsize=(30,10), dpi=150)
plt.plot(df.index,df["Centro Ozono"],lw=2, color='black')
plt.xlabel('Fecha', fontdict = {'fontsize':35 })
plt.ylabel('O3',fontdict = {'fontsize':35 })
plt.title('Ozono en el centro CDMX 2019-2020', fontsize = 35)
plt.tick_params(labelsize=30, pad=10)
```

**pandas.concat**: Concatena o «apila» objetos a lo largo de un eje.

```python
df=pd.concat([c19,c20,c21],axis = 0)
```

```python
df
```

```python
df['Fecha'] = pd.to_datetime(df['Fecha'])
df = df.set_index("Fecha")
```

```python
plt.figure(figsize=(30,10), dpi=150)
plt.plot(df.index,df["Centro Ozono"],lw=2, color='black')
plt.xlabel('Fecha', fontdict = {'fontsize':35 })
plt.ylabel('O3',fontdict = {'fontsize':35 })
plt.title('Ozono en el centro CDMX 2019-2020', fontsize = 35)
plt.tick_params(labelsize=30, pad=10)
```
