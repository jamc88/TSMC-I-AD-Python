```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("rama_2023.csv",parse_dates=True, index_col=0)
```

La concentración de un contaminante del aire se da en microgramos (una millonésima de gramo) por metro cúbico de aire $\mu$g/$m^3$.

Contaminantes registrados en CDMX (2015-01-01 al 31-05-2023):

* ozono (O3) 
* dióxido de azufre (SO2)
* óxido nítrico (NO)
* dióxido de nitrógeno (NO2)
* óxidos de nitrógeno (NOx)
* monóxido de carbono (CO)
* partículas menores a 10 micrómetros (PM10)
* partículas menores a 2.5 micrómetros (PM2.5)
