```python
plt.figure(figsize=(30,10), dpi=150)
plt.plot(df.index,df['NOX'],lw=1,label="monóxido de carbono", color='black')
plt.plot(df.index,df['PM10'],lw=1,label="PM10", color='red')
plt.xlabel('Fecha', fontdict = {'fontsize':35 })
plt.ylabel('NOX',fontdict = {'fontsize':35 }) 
plt.tick_params(labelsize=30, pad=10)
plt.legend(prop={'size':25});
```

```python
from scipy.stats import shapiro

estadistico, p_value = shapiro(df['CO'])
print('Valor del estadístico = %.3f , p_value= %.10f' % (estadistico,p_value)) 
print('\n')

estadistico, p_value = shapiro(df['NO'])
print('Valor del estadístico = %.3f , p_value= %.10f' % (estadistico,p_value)) 
print('\n')

estadistico, p_value = shapiro(df['NO2'])
print('Valor del estadístico = %.3f , p_value= %.10f' % (estadistico,p_value)) 
print('\n')

estadistico, p_value = shapiro(df['NOX'])
print('Valor del estadístico = %.3f , p_value= %.10f' % (estadistico,p_value)) 
print('\n')
```
