```python
df = pd.read_csv('https://raw.githubusercontent.com/jamc88/TSMCI-Analisis-de-datos-con-Python/main/Datos/nacimientos_2021.csv',
                 usecols=['Estado','Regiones','Total'],index_col=['Estado']) 
df
```

De forma predeterminada, pandas.concat trabaja sobre axis='index'

```python
pd.concat([df1,df2,df3])
```

```python
pd.concat([df1,df2,df3], axis='columns')
```

```python
df4 = pd.concat([df1,df3])
df4
```

```python
pd.concat([df1, df4], axis='columns')
```

```python
pd.concat([df1, df4], axis='columns', join='inner')
```

Un posible problema es que las piezas concatenadas no se puedan identificar en el resultado. Supongamos entonces que queremos crear un índice jerárquico en el eje de concatenación. Para ello, empleamos el argumento keys:

```python
result = pd.concat([df1, df2, df3], 
                   keys=['grupo 1', 'grupo 2', 'grupo 3'])
```

```python
result
```

```python
result.unstack()
```

En el caso de combinar con axis='columns', las keys se convierten en los encabezados de columna del dataframe:

```python
pd.concat([df1, df2, df3], axis='columns', 
          keys=['grupo 1', 'grupo 2', 'grupo 3'])
```

Si se pasa un diccionario de objetos en lugar de una lista, se emplearán las claves del diccionario para la opción keys:

```python
pd.concat({'grupo1': df1, 'grupo3':df3},  axis='columns')
```

Hay argumentos adicionales que gestionan el modo en que se crea el índice jerárquico (véase la tabla 8.3). Por ejemplo, podemos asignar nombre a los niveles de eje creados con el argumento names:


```python
pd.concat([df1, df2, df3], axis='columns', 
          keys=['grupo 1', 'grupo 2', 'grupo 3'],
         names=['nivel1','nivel2'])
```

se puede descarta los índices de cada dataframe y concatenar los datos solo de las columnas asignando un nuevo índice por defecto:

```python
pd.concat([df1,df2], ignore_index=True)
```

### Combinar datos con superposición

Hay otra situación de combinación de datos que no se puede expresar como una operación de fusión o concatenación. Es posible que tengamos dos conjuntos de datos con índices que se superponen totalmente o en parte.

```python
df1 = pd.DataFrame({'a': [1., np.nan, 5., np.nan],
                    'b': [np.nan, 2., np.nan, 6.],
                    'c': range(2, 18, 4)})

df2 = pd.DataFrame({'a': [5., 4., np.nan, 3., 7.],
                    'b': [np.nan, 3., 4., 6., 8.]})
```

```python
df2
```

```python
df2 = pd.DataFrame({'a': [5., 4., np.nan, 3., 7.],
                    'b': [np.nan, 3., 4., 6., 8.]})
```

```python
df2
```

Si se desea alinear valores por índice, se utiliza el método **combine_first**, este método lo que hace es
«parchear» los datos que faltan en el objeto de llamada con datos del objeto que se pase:

```python
df1.combine_first(df2)
```

### Remodelación y Transposición

La indexación jerárquica ofrece una forma coherente de reordenar datos en un dataframe. Tenemos dos acciones principales:
* stack: Cambia o transpone las columnas de los datos por las filas.
* unstack: Cambia o transpone las filas por las columnas.


```python
data = pd.DataFrame(np.arange(6).reshape((2, 3)), 
                    index=pd.Index(['Colima', 'Durango'], 
                                   name='Estado'),
                    columns=pd.Index(['Uno', 'Dos', 'Tres'],
                    name='Número'))
                    
```

```python
data
```

Utilizando el método stack con estos datos cambiamos las columnas por las filas, produciendo una serie:

```python
result = data.stack()
```

```python
result
```

A partir de una serie indexada jerárquicamente, se pueden reordenar los datos de nuevo en un dataframe con unstack:

```python
result.unstack(level=0)
```

```python
result.unstack(level='Estado')
```
