# Discretización

Los datos continuos se suelen discretizar o, lo que es lo mismo, separar en «categorías» para su análisis.

```python
corredores = pd.read_csv('Corredores.csv')
```

```python
corredores
```

```python
corredores.describe()
```

Supongamos que queremos agrupar las edades de los corredores en grupos de 5 a 18 años, de 19 a 25 años, de 26 a 35 años, de 36 años a 60 años y de 60 años a 100 años. Para ello tenemos que usar **pandas.cut**:

```python
edades = corredores['Edad']
```

```python
bins = [5,18,25,35,60,100]
```

```python
categorias = pd.cut(edades,bins)
```

```python
categorias
```

**pd.value_counts()** obtiene los recuentos de contenedores para el resultado de pandas.cut.

```python
pd.value_counts(categorias)
```

En la representación de cadena de texto de un intervalo, un paréntesis significa que el lado está abierto, mientras que el corchete significa que está cerrado. Se puede cambiar el lado que está cerrado pasando **right=False**:

```python
pd.cut(edades,bins,right=False)
```

Se puede anular el etiquetado de contenedor predeterminado basado en intervalos pasando una lista o un array a la opción labels:

```python
nombres_grupos = ['Niño','joven_adulto','mediana_edad','Adulto',
                 'tercera_edad']
```

```python
pd.cut(edades,bins,labels=nombres_grupos)
```

Si se pasa un número entero de contenedores a pandas.cut en lugar de bordes explícitos de contenedor, calculará contenedores de la misma longitud basándose en los valores mínimo y máximo de los datos.

```python
data = np.random.standard_normal(size=20)
```

```python
data
```

La opción precision=2 limita la precisión decimal a dos dígitos.

```python
group = pd.cut(data, 4, precision=2)
```

```python
pd.value_counts(group)
```

# Detección y filtrado de valores atípicos

```python
datos = pd.DataFrame(np.random.standard_normal((1000, 4)))
```

```python
datos
```

```python
datos.describe()
```

Supongamos que queremos encontrar valores en una de las columnas que sean mayores de 3 y menores que -3:

```python
col = datos[2]
```

```python
col
```

```python
col[col.abs() > 3]
```

Para seleccionar todas las filas con un valor superior a 3 o –3, se puede emplear el método **any**:

```python
datos[(datos.abs() > 3).any(axis='columns')]
```

Los valores se pueden configurar con base a ciertos criterios. Podemos limitar los valores que quedenn fuera del intervalo –3 a 3:

```python
datos[datos.abs() > 3] = np.sign(datos) * 3
```

```python
datos.iloc[538]
```

La sentencia **np.sign()** produce valores 1 y –1 según si los valores de data son positivos o negativos:

```python
np.sign(datos) 
```
